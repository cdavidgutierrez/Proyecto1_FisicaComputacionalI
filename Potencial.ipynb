```python
%pylab inline
from scipy.integrate import odeint
from scipy.linalg import eigh_tridiagonal
from scipy import constants
```

```python
def Eigen(eff_pot,N=2000):
    """
    Esta función encuentra las autofunciones y autovalores asociados a la ecuación de Schrödinger
    en una dimensión para el potencial "eff_pot". (Ver Documentación).
    """
    
    dy=1/N
    #Calculando el potencial dentro de la región definida por los límites del potencial.
    y=np.linspace(0,1,N+1)
    V=eff_pot(y)
    
    #De acuerdo a lo mostrado en la documentación, los elementos de la diagonal y los de la diagonal secundaria.
    diag=V[1:-1]+1/dy**2
    off_diag=-1/(2*dy**2)*np.ones(len(diag)-1)
    
    #Se resuelve el problema de autovalores matricial.
    E_n,psi_n=eigh_tridiagonal(diag,off_diag)
    return E_n,psi_n.T    
```

```python
#Solución dependiente del tiempo.
def Psi(t,cn,pot):
  """
  Esta función calcula la solución dependiente del tiempo "t" como
  combinación lineal de las primeras 4 autofunciones del potencial entregado.
  """  
  psi=0
  for i in range(0,4):
    En=pot.eigenvalues(i)*constants.hbar**2/(m*l**2)
    psi+=cn[i]*np.exp(-1j*En*t/constants.hbar)*pot.eigenfunctions(i)
  return psi
```

```python
class Potencial():
    """
    ESta clase permite definir potenciales para el problema de la ecuación de Schrödinger.
    """
    def __init__(self,mass,hight,lent,pot_function):
        self.mass=mass
        #Valor constante del potencial fuera de la región definida.
        self.hight=hight
        self.lent=lent
        self.pot_function=pot_function

    def eff_pot(self,x): 
        return self.mass*self.lent**2*self.pot_function(self,x)

    def eigenvalues(self,n):
        """
        Utilizando la función Eigen, devuelve el valor de energía "n" para el potencial.
        """
        return Eigen(self.eff_pot)[0][n]

    def eigenfunctions(self,n):
        """
        Utilizando la función Eigen, devuelve la autofunción asociada al valor de energía "n" para el potencial.
        """
        return Eigen(self.eff_pot)[1][n]
```

```python
m=10
l=10
h=10
```

## Pozo de potencial finito.

```python
def finite_well1(self,x):
    if 0<=x<=1:
      V=0
    else:
      V=10
    return V
finite_well=vectorize(finite_well1)
Finite_well=Potencial(m,h,l,finite_well)
```

```python
plt.title("Autofunciones del pozo finito.")
plt.plot(Finite_well.eigenfunctions(0),label="n=1")
plt.plot(Finite_well.eigenfunctions(1),label="n=2")
plt.plot(Finite_well.eigenfunctions(2),label="n=3")
plt.plot(Finite_well.eigenfunctions(3),label="n=4")
plt.xlabel("y=z/L")
plt.ylabel("$\psi(y)$")
plt.legend()
plt.grid()
```

```python
finWellEn=[Finite_well.eigenvalues(0),Finite_well.eigenvalues(1),Finite_well.eigenvalues(2),Finite_well.eigenvalues(3)]
plt.bar(np.arange(0,4,1),finWellEn)
plt.ylabel("$mL^2E_n/\hbar^2$")
```

```python
plt.title("Densidades de probabilidad del pozo finito.")
plt.plot(Finite_well.eigenfunctions(0)**2,label="n=1")
plt.plot(Finite_well.eigenfunctions(1)**2,label="n=2")
plt.plot(Finite_well.eigenfunctions(2)**2,label="n=3")
plt.plot(Finite_well.eigenfunctions(3)**2,label="n=4")
plt.legend(loc=1)
plt.xlabel("y/L")
plt.ylabel("$|\psi(y)|^2$")
plt.grid()
```
